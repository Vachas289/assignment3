```python
def find_outliers(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]

# Detect outliers for each column
outliers_marks = find_outliers(df, 'Overall_Marks')
outliers_income = find_outliers(df, 'Monthly_Family_Income')
outliers_age = find_outliers(df, 'Age')

# Print the outliers
print("Outliers in Overall Marks:\n", outliers_marks)
print("\nOutliers in Family Income:\n", outliers_income)
print("\nOutliers in Age:\n", outliers_age)
```

```python
def assign_grade(marks):
    percentage = (marks / 500) * 100  
    if marks >= 450:
        return 'A'
    elif marks >= 400:
        return 'B'
    elif marks >= 350:
        return 'C'
    elif marks >= 300:
        return 'D'
    elif marks >= 200:
        return 'F (Pass)'
    else:
        return 'F (Fail)'

df['Grade'] = df['Overall_Marks'].apply(assign_grade)

print(df[['Student_ID', 'Overall_Marks', 'Grade']])
```

```python
from scipy import stats

montfort_marks = df[df['School_Name'] == 'Montfort']['Overall_Marks']
rla_marks = df[df['School_Name'] == 'RLA']['Overall_Marks']

t_stat, p_value = stats.ttest_ind(montfort_marks, rla_marks)

null_hypothesis = "The mean overall marks of students from Montfort are equal to the mean overall marks of students from RLA."
alternative_hypothesis = "The mean overall marks of students from Montfort are not equal to the mean overall marks of students from RLA."

if p_value < 0.05:
    conclusion = "Reject the null hypothesis: There is a significant difference in the overall marks between Montfort and RLA."
else:
    conclusion = "Fail to reject the null hypothesis: There is no significant difference in the overall marks between Montfort and RLA."

print(f"Null Hypothesis: {null_hypothesis}")
print(f"Alternative Hypothesis: {alternative_hypothesis}")
print(f"T-statistic: {t_stat}, P-value: {p_value}")
print(f"Conclusion: {conclusion}")
```

```python
class_marks = [df[df['Class'] == class_]['Overall_Marks'] for class_ in df['Class'].unique()]

f_stat, p_value = stats.f_oneway(*class_marks)

null_hypothesis = "The mean overall marks of students in all classes are equal."
alternative_hypothesis = "At least one class has a mean overall mark that differs from the others."

if p_value < 0.05:
    conclusion = "Reject the null hypothesis: There is a significant difference in the overall marks across the classes."
else:
    conclusion = "Fail to reject the null hypothesis: There is no significant difference in the overall marks across the classes."
print(f"Null Hypothesis: {null_hypothesis}")
print(f"Alternative Hypothesis: {alternative_hypothesis}")
print(f"F-statistic: {f_stat}, P-value: {p_value}")
print(f"Conclusion: {conclusion}")
```

```python
corr, p_value = stats.pearsonr(df['Monthly_Family_Income'], df['Overall_Marks'])

null_hypothesis = "There is no correlation between family income and overall marks."
alternative_hypothesis = "There is a correlation between family income and overall marks."

if p_value < 0.05:
    conclusion = "Reject the null hypothesis: There is a significant correlation between family income and overall marks."
else:
    conclusion = "Fail to reject the null hypothesis: There is no significant correlation between family income and overall marks."

print(f"Null Hypothesis: {null_hypothesis}")
print(f"Alternative Hypothesis: {alternative_hypothesis}")
print(f"Pearson correlation: {corr}, P-value: {p_value}")
print(f"Conclusion: {conclusion}")
```

```python
contingency_table = pd.crosstab(df['Class'], df['Can_Afford'])
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

null_hypothesis = "There is no relationship between class and the ability to afford the trip."
alternative_hypothesis = "There is a relationship between class and the ability to afford the trip."

if p_value < 0.05:
    conclusion = "Reject the null hypothesis: There is a significant relationship between class and the ability to afford the trip."
else:
    conclusion = "Fail to reject the null hypothesis: There is no significant relationship between class and the ability to afford the trip."

print(f"Null Hypothesis: {null_hypothesis}")
print(f"Alternative Hypothesis: {alternative_hypothesis}")
print(f"Chi2-statistic: {chi2_stat}, P-value: {p_value}")
print(f"Conclusion: {conclusion}")
```

```python
group_1 = df[df['Age'] <= 10]['Overall_Marks']
group_2 = df[df['Age'] > 10]['Overall_Marks']

t_stat, p_value = stats.ttest_ind(group_1, group_2)

null_hypothesis = "The mean overall marks of students aged 10 or less are equal to the mean overall marks of students aged greater than 10."
alternative_hypothesis = "The mean overall marks of students aged 10 or less are not equal to the mean overall marks of students aged greater than 10."

if p_value < 0.05:
    conclusion = "Reject the null hypothesis: There is a significant difference in the overall marks between the two age groups."
else:
    conclusion = "Fail to reject the null hypothesis: There is no significant difference in the overall marks between the two age groups."

print(f"Null Hypothesis: {null_hypothesis}")
print(f"Alternative Hypothesis: {alternative_hypothesis}")
print(f"T-statistic: {t_stat}, P-value: {p_value}")
print(f"Conclusion: {conclusion}")
```
